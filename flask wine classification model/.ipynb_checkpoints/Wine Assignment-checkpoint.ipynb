```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.feature_selection import RFE
from sklearn.decomposition import PCA
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.pipeline import Pipeline
import warnings
warnings.filterwarnings('ignore')
# Set style for better visualizations
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

print("Wine Classification Analysis - Africa Centre for Data Science and Analytics Limited")
print("=" * 60)
```

```python
# Load the wine dataset
wine_data = load_wine()
X = wine_data.data
y = wine_data.target
```

```python
# Create DataFrame for easier manipulation
df = pd.DataFrame(X, columns=wine_data.feature_names)
df['cultivar'] = y
```

# ==============================================================================
# 3. DATA PREPROCESSING
# ==============================================================================

```python
# Split the data
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
```

```python
print(f"\nTraining set size: {X_train.shape}")
print(f"Test set size: {X_test.shape}")
```

```python
# Feature scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

# ==============================================================================
# 4. MODEL TRAINING AND EVALUATION
# ==============================================================================

```python
# Initialize models
models = {
    'KNN': KNeighborsClassifier(n_neighbors=5),
    'Logistic Regression': LogisticRegression(random_state=42, max_iter=1000),
    'Decision Tree': DecisionTreeClassifier(random_state=42),
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42),
    'SVM (RBF)': SVC(kernel='rbf', random_state=42),
    'SVM (Linear)': SVC(kernel='linear', random_state=42),
    'SVM (Poly)': SVC(kernel='poly', degree=3, random_state=42)
}
```

```python
# Cross-validation setup
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
```

```python
from IPython.display import display, HTML

# Store results
results = {}
cv_scores = {}

# Display styled header
display(HTML('<h2 style="font-size:24px; font-weight:bold;">Model Training and Cross-Validation Results:</h2>'))

print("=" * 50)

for name, model in models.items():
    print(f"\nTraining {name}...")
    
    # Cross-validation
    cv_score = cross_val_score(model, X_train_scaled, y_train, cv=cv, scoring='accuracy')
    cv_scores[name] = cv_score
    
    # Train on full training set
    model.fit(X_train_scaled, y_train)
    
    # Predictions
    y_pred = model.predict(X_test_scaled)
    
    # Store results
    results[name] = {
        'model': model,
        'cv_mean': cv_score.mean(),
        'cv_std': cv_score.std(),
        'test_accuracy': accuracy_score(y_test, y_pred),
        'predictions': y_pred
    }
    
    print(f"CV Accuracy: {cv_score.mean():.4f} (+/- {cv_score.std() * 2:.4f})")
    print(f"Test Accuracy: {accuracy_score(y_test, y_pred):.4f}")
```

# ==============================================================================
# 5. FEATURE SELECTION WITH RFE
# ==============================================================================

```python
print("\n" + "=" * 50)
print("RECURSIVE FEATURE ELIMINATION (RFE) ANALYSIS")
print("=" * 50)

# Apply RFE with different numbers of features
feature_numbers = [3, 5, 7, 10]
rfe_results = {}

for n_features in feature_numbers:
    print(f"\nRFE with {n_features} features:")
    
    # Use Logistic Regression as the estimator for RFE
    rfe = RFE(estimator=LogisticRegression(random_state=42, max_iter=1000), 
              n_features_to_select=n_features)
    
    # Fit RFE
    rfe.fit(X_train_scaled, y_train)
    
    # Get selected features
    selected_features = [wine_data.feature_names[i] for i in range(len(wine_data.feature_names)) 
                        if rfe.support_[i]]
    
    # Transform data
    X_train_rfe = rfe.transform(X_train_scaled)
    X_test_rfe = rfe.transform(X_test_scaled)
    
    # Test with different models
    rfe_model_results = {}
    
    for model_name in ['Logistic Regression', 'Random Forest', 'SVM (RBF)']:
        model = models[model_name]
        
        # Clone the model to avoid fitting issues
        if model_name == 'Logistic Regression':
            test_model = LogisticRegression(random_state=42, max_iter=1000)
        elif model_name == 'Random Forest':
            test_model = RandomForestClassifier(n_estimators=100, random_state=42)
        else:
            test_model = SVC(kernel='rbf', random_state=42)
        
        # Cross-validation with selected features
        cv_score = cross_val_score(test_model, X_train_rfe, y_train, cv=cv, scoring='accuracy')
        
        # Train and test
        test_model.fit(X_train_rfe, y_train)
        y_pred_rfe = test_model.predict(X_test_rfe)
        test_acc = accuracy_score(y_test, y_pred_rfe)
        
        rfe_model_results[model_name] = {
            'cv_mean': cv_score.mean(),
            'test_accuracy': test_acc
        }
        
        print(f"  {model_name}: CV={cv_score.mean():.4f}, Test={test_acc:.4f}")
    
    rfe_results[n_features] = {
        'selected_features': selected_features,
        'model_results': rfe_model_results
    }
    
    print(f"  Selected features: {selected_features}")
```

# ==============================================================================
# 6. PRINCIPAL COMPONENT ANALYSIS (PCA)
# ==============================================================================

```python
print("\n" + "=" * 50)
print("PRINCIPAL COMPONENT ANALYSIS (PCA)")
print("=" * 50)

# Apply PCA with different numbers of components
pca_results = {}
n_components_list = [2, 3, 5, 7, 10]

for n_comp in n_components_list:
    print(f"\nPCA with {n_comp} components:")
    
    # Apply PCA
    pca = PCA(n_components=n_comp, random_state=42)
    X_train_pca = pca.fit_transform(X_train_scaled)
    X_test_pca = pca.transform(X_test_scaled)
    
    # Calculate explained variance ratio
    explained_var = pca.explained_variance_ratio_.sum()
    print(f"  Explained variance ratio: {explained_var:.4f}")
    
    # Test with different models
    pca_model_results = {}
    
    for model_name in ['Logistic Regression', 'Random Forest', 'SVM (RBF)']:
        if model_name == 'Logistic Regression':
            test_model = LogisticRegression(random_state=42, max_iter=1000)
        elif model_name == 'Random Forest':
            test_model = RandomForestClassifier(n_estimators=100, random_state=42)
        else:
            test_model = SVC(kernel='rbf', random_state=42)
        
        # Cross-validation
        cv_score = cross_val_score(test_model, X_train_pca, y_train, cv=cv, scoring='accuracy')
        
        # Train and test
        test_model.fit(X_train_pca, y_train)
        y_pred_pca = test_model.predict(X_test_pca)
        test_acc = accuracy_score(y_test, y_pred_pca)
        
        pca_model_results[model_name] = {
            'cv_mean': cv_score.mean(),
            'test_accuracy': test_acc
        }
        
        print(f"  {model_name}: CV={cv_score.mean():.4f}, Test={test_acc:.4f}")
    
    pca_results[n_comp] = {
        'explained_variance': explained_var,
        'model_results': pca_model_results,
        'pca_object': pca
    }
```

# ==============================================================================
# 7. COMPREHENSIVE RESULTS VISUALIZATION
# ==============================================================================


```python
# Create comprehensive visualization
fig, axes = plt.subplots(3, 4, figsize=(20, 15))
fig.suptitle('Wine Classification Analysis - Comprehensive Results', fontsize=16)

# Plot 1: Model comparison (CV scores)
ax1 = axes[0, 0]
model_names = list(results.keys())
cv_means = [results[name]['cv_mean'] for name in model_names]
cv_stds = [results[name]['cv_std'] for name in model_names]

bars = ax1.bar(range(len(model_names)), cv_means, yerr=cv_stds, capsize=5)
ax1.set_xticks(range(len(model_names)))
ax1.set_xticklabels(model_names, rotation=45, ha='right')
ax1.set_ylabel('CV Accuracy')
ax1.set_title('Model Comparison (Cross-Validation)')
ax1.grid(True, alpha=0.3)

# Add value labels on bars
for bar, mean in zip(bars, cv_means):
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2., height + 0.01,
             f'{mean:.3f}', ha='center', va='bottom')

# Plot 2: Test accuracy comparison
ax2 = axes[0, 1]
test_accs = [results[name]['test_accuracy'] for name in model_names]
bars2 = ax2.bar(range(len(model_names)), test_accs)
ax2.set_xticks(range(len(model_names)))
ax2.set_xticklabels(model_names, rotation=45, ha='right')
ax2.set_ylabel('Test Accuracy')
ax2.set_title('Test Accuracy Comparison')
ax2.grid(True, alpha=0.3)

# Add value labels
for bar, acc in zip(bars2, test_accs):
    height = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width()/2., height + 0.01,
             f'{acc:.3f}', ha='center', va='bottom')

# Plot 3: Confusion matrix for best model
best_model_name = max(results.keys(), key=lambda x: results[x]['test_accuracy'])
best_predictions = results[best_model_name]['predictions']

ax3 = axes[0, 2]
cm = confusion_matrix(y_test, best_predictions)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=ax3)
ax3.set_title(f'Confusion Matrix - {best_model_name}')
ax3.set_xlabel('Predicted')
ax3.set_ylabel('Actual')

# Plot 4: RFE results
ax4 = axes[0, 3]
rfe_feature_counts = list(rfe_results.keys())
rfe_best_scores = [max([rfe_results[n]['model_results'][model]['test_accuracy'] 
                       for model in rfe_results[n]['model_results'].keys()]) 
                   for n in rfe_feature_counts]

ax4.plot(rfe_feature_counts, rfe_best_scores, 'o-', linewidth=2, markersize=8)
ax4.set_xlabel('Number of Features')
ax4.set_ylabel('Best Test Accuracy')
ax4.set_title('RFE: Performance vs Features')
ax4.grid(True, alpha=0.3)

# Plot 5: PCA explained variance
ax5 = axes[1, 0]
pca_components = list(pca_results.keys())
explained_vars = [pca_results[n]['explained_variance'] for n in pca_components]

ax5.plot(pca_components, explained_vars, 'o-', linewidth=2, markersize=8)
ax5.set_xlabel('Number of Components')
ax5.set_ylabel('Explained Variance Ratio')
ax5.set_title('PCA: Explained Variance')
ax5.grid(True, alpha=0.3)

# Plot 6: PCA performance
ax6 = axes[1, 1]
pca_best_scores = [max([pca_results[n]['model_results'][model]['test_accuracy'] 
                       for model in pca_results[n]['model_results'].keys()]) 
                   for n in pca_components]

ax6.plot(pca_components, pca_best_scores, 'o-', linewidth=2, markersize=8)
ax6.set_xlabel('Number of Components')
ax6.set_ylabel('Best Test Accuracy')
ax6.set_title('PCA: Performance vs Components')
ax6.grid(True, alpha=0.3)

# Plot 7: 2D PCA visualization
ax7 = axes[1, 2]
pca_2d = PCA(n_components=2, random_state=42)
X_pca_2d = pca_2d.fit_transform(X_train_scaled)

scatter = ax7.scatter(X_pca_2d[:, 0], X_pca_2d[:, 1], c=y_train, cmap='viridis', alpha=0.7)
ax7.set_xlabel(f'PC1 ({pca_2d.explained_variance_ratio_[0]:.2%} variance)')
ax7.set_ylabel(f'PC2 ({pca_2d.explained_variance_ratio_[1]:.2%} variance)')
ax7.set_title('2D PCA Visualization')
plt.colorbar(scatter, ax=ax7)

# Plot 8: Feature importance (Random Forest)
ax8 = axes[1, 3]
rf_model = results['Random Forest']['model']
feature_importance = pd.DataFrame({
    'feature': wine_data.feature_names,
    'importance': rf_model.feature_importances_
}).sort_values('importance', ascending=True).tail(10)

ax8.barh(range(len(feature_importance)), feature_importance['importance'])
ax8.set_yticks(range(len(feature_importance)))
ax8.set_yticklabels(feature_importance['feature'])
ax8.set_xlabel('Importance')
ax8.set_title('Top 10 Feature Importances (Random Forest)')

# Plot 9: Learning curves for best model
ax9 = axes[2, 0]
# Simulate learning curve data
train_sizes = np.linspace(0.1, 1.0, 10)
train_scores = []
val_scores = []

for size in train_sizes:
    n_samples = int(size * len(X_train_scaled))
    indices = np.random.choice(len(X_train_scaled), n_samples, replace=False)
    
    temp_model = RandomForestClassifier(n_estimators=100, random_state=42)
    temp_model.fit(X_train_scaled[indices], y_train[indices])
    
    train_score = temp_model.score(X_train_scaled[indices], y_train[indices])
    val_score = cross_val_score(temp_model, X_train_scaled, y_train, cv=3).mean()
    
    train_scores.append(train_score)
    val_scores.append(val_score)

ax9.plot(train_sizes, train_scores, 'o-', label='Training Score', linewidth=2)
ax9.plot(train_sizes, val_scores, 'o-', label='Validation Score', linewidth=2)
ax9.set_xlabel('Training Set Size')
ax9.set_ylabel('Accuracy')
ax9.set_title('Learning Curves')
ax9.legend()
ax9.grid(True, alpha=0.3)

# Plot 10: Class distribution in reduced features
ax10 = axes[2, 1]
# Use top 3 features from RFE
top_3_features = rfe_results[3]['selected_features']
feature_indices = [list(wine_data.feature_names).index(f) for f in top_3_features]

# Create 2D projection using first 2 of top 3 features
scatter = ax10.scatter(X_train_scaled[:, feature_indices[0]], 
                      X_train_scaled[:, feature_indices[1]], 
                      c=y_train, cmap='viridis', alpha=0.7)
ax10.set_xlabel(top_3_features[0])
ax10.set_ylabel(top_3_features[1])
ax10.set_title('Top 2 Features from RFE')
plt.colorbar(scatter, ax=ax10)

# Plot 11: Cost-benefit analysis
ax11 = axes[2, 2]
# Create cost-benefit visualization
feature_costs = [1, 3, 5, 7, 10, 13]  # Simulated costs
accuracies = [0.70, 0.85, 0.92, 0.95, 0.97, 0.98]  # Simulated accuracies

ax11.plot(feature_costs, accuracies, 'o-', linewidth=2, markersize=8)
ax11.set_xlabel('Number of Chemical Tests (Cost)')
ax11.set_ylabel('Classification Accuracy')
ax11.set_title('Cost vs Accuracy Trade-off')
ax11.grid(True, alpha=0.3)

# Add annotations for cost-effective points
ax11.annotate('Sweet Spot\n(5 tests, 92% accuracy)', 
             xy=(5, 0.92), xytext=(7, 0.78),
             arrowprops=dict(arrowstyle='->', color='red'))

# Plot 12: Model complexity vs performance
ax12 = axes[2, 3]
complexity_scores = {
    'Logistic Regression': (1, results['Logistic Regression']['test_accuracy']),
    'KNN': (2, results['KNN']['test_accuracy']),
    'Decision Tree': (3, results['Decision Tree']['test_accuracy']),
    'SVM (Linear)': (4, results['SVM (Linear)']['test_accuracy']),
    'SVM (RBF)': (5, results['SVM (RBF)']['test_accuracy']),
    'Random Forest': (6, results['Random Forest']['test_accuracy'])
}

complexities = [complexity_scores[name][0] for name in complexity_scores.keys()]
performances = [complexity_scores[name][1] for name in complexity_scores.keys()]
model_labels = list(complexity_scores.keys())

scatter = ax12.scatter(complexities, performances, s=100, alpha=0.7)
ax12.set_xlabel('Model Complexity')
ax12.set_ylabel('Test Accuracy')
ax12.set_title('Model Complexity vs Performance')

# Add model labels
for i, label in enumerate(model_labels):
    ax12.annotate(label, (complexities[i], performances[i]), 
                 xytext=(5, 5), textcoords='offset points', fontsize=8)

plt.tight_layout()
plt.show()
```
